```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# parquet_path = "/depot/cms/hmm/shar1172/hmm_ntuples/skimmed_for_dnn/2016postVFP/vbf_powheg_dipole/part0.parquet"
parquet_path = "/depot/cms/hmm/shar1172/hmm_ntuples/skimmed_for_dnn/2016postVFP/ttjets_dl/part2.parquet"
col = ["ll_zstar_log_nominal", "pt_centrality_nominal", "zeppenfeld_nominal"]

if not os.path.exists(parquet_path):
    print("File not found:", parquet_path)
    sys.exit(1)
```

```python
# Read only the requested column to save memory
try:
    df = pd.read_parquet(parquet_path, columns=col)
except Exception as e:
    print("Failed to read parquet or column. Error:", e)
    # try to list available columns for debugging
    try:
        tmp = pd.read_parquet(parquet_path)
        print("Available columns:", list(tmp.columns))
    except Exception:
        pass
    sys.exit(1)
```

```python
df.head(50)
```

```python
df.fillna(-99.0, inplace=True)
df.head(50)
```

```python
vals = df
# vals = df[col].dropna().values
# print(f"Loaded {len(vals)} entries for {col}; mean={np.mean(vals):.4g}, std={np.std(vals):.4g}")

for column in col:
    vals = df[column]
    plt.figure(figsize=(8,5))
    plt.hist(vals, bins=800, histtype="stepfilled", alpha=0.7, color="C0")
    plt.xlim(-100,100)
    plt.xlabel(column)
    plt.ylabel("Entries")
    plt.title(f"{column} (part0.parquet)")
    plt.grid(alpha=0.3)
    out_png = f"{column}.png"
    plt.tight_layout()
    plt.savefig(out_png)
    print(f"Saved histogram to {out_png}")
    plt.show()
```
